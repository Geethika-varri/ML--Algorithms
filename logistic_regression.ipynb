```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv("Loan Eligibility Prediction.csv")
df
```

```python
# converting categorical to numerical
df['Gender'] = df['Gender'].map({"Male":0,"Female" :1})
df['Gender']
```

```python
df['Married'] = df['Married'].map({"No":0,"Yes" :1})
df['Married']
```

```python
df['Education'] = df['Education'].map({"Not Graduate":0,"Graduate" :1})
df['Education']
```

```python
df['Self_Employed'] = df['Self_Employed'].map({"No":0,"Yes" :1})
df['Self_Employed'] 
```

```python
df['Loan_Status'] = df['Loan_Status'].map({"N":0, "Y" :1})
df['Loan_Status'] 
```

```python
df
```

```python
new_df = df.drop(['Customer_ID', 'Dependents', 'Coapplicant_Income', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Loan_Status'], axis=1)
new_df
```

```python
Loan_Status = df.Loan_Status
Loan_Status
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_df,Loan_Status,train_size = 0.8)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
model
```

```python
y_test
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
model.score(x_test,y_test)
```
